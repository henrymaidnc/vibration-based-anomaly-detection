# Day 1 â€“ Raw Vibration Signal Inspection

Objective:
- Understand the structure of the IMS vibration data
- Observe raw vibration signals in the time domain
- Identify early visual signs of bearing degradation

```python
import pandas as pd

file_path = "../ims_dataset/1st_test/2003.10.22.12.06.24"
data = pd.read_csv(file_path, sep="\t", header=None)
print("Dataset shape:", data.shape)
display(data.head())
```

Observation:
The dataset contains 20480 samples with 8 vibration channels,
indicating a multi-sensor industrial measurement setup.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,4))
plt.plot(data.iloc[:, 0])
plt.title("Raw Vibration Signal â€“ Channel 1")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()
```

Observation:
- The vibration signal appears noisy and oscillatory,
suggesting continuous mechanical motion rather than random noise.
- There are intermittent increases in vibration amplitude, which may indicate transient mechanical events or early-stage irregularities.

Conclusion:
- The vibration signal exhibits continuous oscillatory behavior centered around zero,
which is consistent with mechanical vibration measured by an accelerometer.

- The amplitude remains relatively small, indicating normal operational conditions,
with occasional higher-amplitude segments that may correspond to transient mechanical events.

-  At this stage, no fault diagnosis can be made.
However, the observed signal characteristics justify further analysis in the frequency domain.
